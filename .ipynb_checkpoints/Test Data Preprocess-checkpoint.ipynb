```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from scipy import stats
from sklearn import metrics
from sklearn import preprocessing
import category_encoders as ce
import pickle
import datetime
import statistics
from sklearn.neighbors import KNeighborsRegressor
```

```python
test = pd.read_csv('Data\Test.csv', index_col=0)
test.head()
```

```python
test[['shipping_charges', 'payment_emi_years(max)']] = test[['shipping_charges', 'payment_emi_years(max)']].fillna(0)
test['X_1'] = test['X_1'].fillna(statistics.mode(test['X_1']))
test['X_2'] = test['X_2'].fillna(statistics.mode(test['X_2']))
test['product_quality_type'] = test['product_quality_type'].fillna('Unknown')
test['product_category_code'] = test['product_category_code'].fillna(-1)
reference_date = pickle.load(open('Data//reference_date.sav', 'rb'))
test['delivery_eta'] = test['delivery_eta'].apply(lambda x:datetime.datetime.strptime(x, '%Y-%m-%d %H:%M:%S'))
test['order_checkout_date'] = test['order_checkout_date'].apply(lambda x:datetime.datetime.strptime(x, '%Y-%m-%d %H:%M:%S'))
test['wait-time'] = (test['delivery_eta'] - test['order_checkout_date']).apply(lambda x: x.total_seconds()/3600)
test['delivery_eta'] = test['delivery_eta'].apply(lambda x:(reference_date-x).total_seconds()/3600)
test['order_checkout_date'] = test['order_checkout_date'].apply(lambda x:(reference_date-x).total_seconds()/3600)
test['volume'] = test['product_height']*test['product_length']*test['product_breadth']
test['product_density'] = test['product_mass']/test['volume']
qualityEnc = pickle.load(open('Data\\qualityEnc.sav', 'rb'))
test.loc[test['product_quality_type']=='O', 'product_quality_type'] = 'Unknown'
test['product_quality_type'] = 12-qualityEnc.transform(test['product_quality_type'])
test.head()
```

```python
def fixNaNPrices(df):
    naInd = df.isna().any(axis=1)

    scaler = preprocessing.StandardScaler()
    X = scaler.fit_transform(df[~naInd].drop(['price'], axis=1))
    y = df[~naInd]['price']

    X_test = scaler.transform(df[naInd].drop(['price'], axis=1))

    neigh = KNeighborsRegressor()
    neigh.fit(X, y)
    df.loc[naInd, 'price'] = neigh.predict(X_test)
    df['price_per_g'] = df['price']/df['product_mass']
    df['price_per_vol'] = df['price']/df['volume']
```

```python
catCols = ['seller_id', 'product_category_code', 'order_status', 'transaction_type']
data = test.copy()
```

#### Frequency Encoding

```python
testFreq = data.copy()

def freqEnc(df, feature):
    freq = df.groupby(feature).size()/len(df)
    df[feature] = df[feature].map(freq)
```

```python
for catCol in catCols:
    freqEnc(testFreq, catCol)
    
fixNaNPrices(testFreq)
pickle.dump(testFreq, open('Data\\testFreq.sav', 'wb'))
```

#### Target Encoding

```python
targEncs = pickle.load(open('Data\\targEncs.sav', 'rb'))
testTarg = data.copy()

for outCol in targEncs:
    targEnc = targEncs[outCol]
    testTarg[[col + ':' + outCol for col in catCols]] = targEnc.transform(testTarg[catCols])
    
testTarg.drop(catCols, axis=1, inplace=True)
fixNaNPrices(testTarg)
pickle.dump(testTarg, open('Data\\testTarg.sav', 'wb'))
```

#### Weight of Evidence Encoding

```python
woeEncs = pickle.load(open('Data\\woeEncs.sav', 'rb'))
testWoe = data.copy()

for outCol in woeEncs:
    woeEnc = woeEncs[outCol]
    testWoe[[col + ':' + outCol for col in catCols]] = woeEnc.transform(testWoe[catCols])
    
testWoe.drop(catCols, axis=1, inplace=True)
fixNaNPrices(testWoe)
pickle.dump(testWoe, open('Data\\testWoe.sav', 'wb'))
```

#### James Stein Encoder

```python
jsEncs = pickle.load(open('Data\\jsEncs.sav', 'rb'))
testJs = data.copy()

for outCol in jsEncs:
    jsEnc = jsEncs[outCol]
    testJs[[col + ':' + outCol for col in catCols]] = jsEnc.transform(testJs[catCols])
    
testJs.drop(catCols, axis=1, inplace=True)
fixNaNPrices(testJs)
pickle.dump(testJs, open('Data\\testJs.sav', 'wb'))
```

```python
meEncs = pickle.load(open('Data\\meEncs.sav', 'rb'))
testMe = data.copy()

for outCol in meEncs:
    meEnc = meEncs[outCol]
    testMe[[col + ':' + outCol for col in catCols]] = meEnc.transform(testMe[catCols])
    
testMe.drop(catCols, axis=1, inplace=True)
fixNaNPrices(testMe)
pickle.dump(testMe, open('Data\\testMe.sav', 'wb'))
```
